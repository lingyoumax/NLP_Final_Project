```python
import matplotlib.pyplot as plt
from getData import getData
from config import GLOBAL_CONFIG

models=GLOBAL_CONFIG['models']
```

# Stability

```python
data = getData("Data/eveluateResults_stability.json")

# 分别获取键和值，用于条形图的x和height参数
keys = data.keys()
values = data.values()

# 创建条形图
bars = plt.bar(keys, values)

# 添加标题和轴标签
plt.title('Model Stability Scores Distribution')
plt.xlabel('Model Name')
plt.ylabel('Stability Score')
plt.show()
```
